```python
class Student:
    def __init__(self, name,surename, roll_number, hindi_marks,english_marks):
        self.name = name
        self.roll_number = roll_number
        self.surename = surename
        self.hindi_marks = hindi_marks
        self.english_marks = english_marks
class StudentBodyCouncil:
    def __init__(self):
        self.students = []

    def accept_student(self):
        name = input("student name: ")
        roll_number = input(" student roll number: ")
        surename = input("student surename: ")
        hindi_marks = input(" student hindi_marks: ")
        english_marks = input("student english_marks: ")
        student = Student(name,surename, roll_number, hindi_marks, english_marks)
        self.students.append(student)
        print(f"Student '{name}' successful!")

    def search_student(self):
        roll_number = input(" student roll number: ")
        for student in self.students:
            if student.roll_number == roll_number:
                print("Student found:")
                print(f"Name: {student.name}")
                print(f"surename: {student.surename}")
                print(f"Roll Number: {student.roll_number}")
                print(f"hindi_marks: {student.hindi_marks}")
                print(f"english_marks: {student.english_marks}")
                return
        print(f"Student with roll number '{roll_number}' not occured.")

    def remove_student(self):
        roll_number = input(" student roll number: ")
        for student in self.students:
            if student.roll_number == roll_number:
                self.students.remove(student)
                print(f"Student with roll number '{roll_number}' removed successfully!")
                return
        print(f"Student with roll number '{roll_number}' not occured.")

    def update_student(self):
        roll_number = input(" student roll number: ")
        for student in self.students:
            if student.roll_number == roll_number:
                print("Enter new student details:")
                name = input("Name: ")
                surename = input("surename: ")
                hindi_marks = input("hindi_marks: ")
                english_marks = input("english_marks: ")
                student.name = name
                student.surename = surename
                student.hindi_marks = hindi_marks
                student.english_marks = english_marks
                print(f"Student with roll number '{roll_number}' successfull!")
                return
        print(f"Student with roll number '{roll_number}' not occured.")

    def display_students(self):
        if not self.students:
            print("No students occured.")
            return

        print("Student List:")
        print("--------------")
        for student in self.students:
            print(f"Name: {student.name}")
            print(f"surename:{student.surename}")
            print(f"Roll Number: {student.roll_number}")
            print(f"hindi_marks: {student.hindi_marks}")
            print(f"english_marks: {student.english_marks}")
            print("--------------")


obj = StudentBodyCouncil()

while True:
    print("\nSTUDENT BODY CoUNCIL")
    print("------------------------")
    print("1. accept ")
    print("2. update ")
    print("3. search ")
    print("4. remove ")
    print("5. Display ")
    print("6. Exit")

    choice = input("Enter your preference (1-6): ")

    if choice == "1":
        obj.accept_student()
    elif choice == "2":
        obj.update_student()
    elif choice == "3":
        obj.search_student()
    elif choice == "4":
        obj.remove_student()
    elif choice == "5":
        obj.display_students()
    elif choice == "6":
        print("Exiting...")
        break
    else:
        print("not found. Please try again.")
```
